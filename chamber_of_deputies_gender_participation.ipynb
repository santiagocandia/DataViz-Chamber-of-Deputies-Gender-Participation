```python
# Importing the .csv file from Github
filepath = ('https://raw.githubusercontent.com/santiagocandia/DataViz-Diputados/main/diputados1.9.csv')
df = pd.read_csv(filepath)
df
```

```python
# Droping columns I donÂ´t need
df = df.drop(['diputado_apellido', 'diputado_nombre',  'diputado_distrito' , 'mandato_fin', 'diputado_juramento', 'diputado_bloque', 'bloque_inicio', 'bloque_fin'], axis =1)
df
```

```python
# Renaming columns
df = df.rename (columns = {'diputado_id': '_id', 'diputado_genero': 'gender', 'mandato_inicio' : 'start', 'diputado_cese':'end'})
df
```

```python
#Keeping year value
df['start'] = df['start'].str[:4]
df['end'] = df['end'].str[:4]
df
```

```python
# Setting values as string and integer
df['_id'] = df['_id'].astype(str)
df['gender'] = df['gender'].astype(str)
df['start'] = df['start'].astype(int)
df['end'] = df['end'].astype(int)
df
```

```python
df['periods'] = df['end'] - df['start']
df
```

```python
df['periods'] = df['periods'].replace(to_replace = 0, value = 1)
df
```

```python
# Bulding fuction to transform de original data frame to a new one with an apropiette structure to work in Tableau
def apply_transform_id(name:str, times:int):
    return [name]*times

def apply_transform_gender(name:str, times:int):
    return [name]*times

def apply_transform_dates(date: int, times:int):
    return pd.date_range(str(int(date)+1), periods=times, freq = 'Y').year

def apply_all(df):
    new_df = pd.DataFrame()
    new_df['_id'] = (df.apply(lambda x: apply_transform_id(x['_id'], x['periods']), axis=1)).explode()
    new_df['gender'] = (df.apply(lambda x: apply_transform_gender(x['gender'], x['periods']), axis=1)).explode()
    new_df['years'] = (df.apply(lambda x: apply_transform_dates(x['start'], x['periods']), axis=1)).explode()
    return new_df

new_df = apply_all(df)
new_df
```
